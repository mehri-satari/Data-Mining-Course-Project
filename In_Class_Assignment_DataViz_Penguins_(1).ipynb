## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.xlabel('bill_length_mm(jittered)')
plt.ylabel('bill_depth_mm(jittered)')
plt.show()
```

**Your interpretation (2–3 sentences):**

_Write your observations here._

```python
#Adelie's have shorter bills but more bill depth, while Gentoo's have longer bills and less bill depth. Chinstrap's have both longer bills and bill depth.
```

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

bin = sns.histplot(data=penguins, x='bill_length_mm', bins=10)
plt.hist(penguins['bill_length_mm'], bins=10)
plt.xlabel('bill_length_mm')
plt.ylabel('Frequency')
plt.show()
```

```python
# TODO: Histogram with 30 bins
bin = sns.histplot(data=penguins, x='bill_length_mm', bins=30)
plt.hist(penguins['bill_length_mm'], bins=30)
plt.xlabel('bill_length_mm')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._

```python
#More bins provides more data. When you increase the bin number, you also increase the dataset, this means not only is there more to visualize, but also more accurate information to intake. On the other hand, less bins create a more simple and clean graph.
```

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
