```python
# ==== QUESTION 1: SENTIMENT ANALYSIS  =======================

# 1. Imports & Setup
import warnings
warnings.filterwarnings("ignore")

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

from wordcloud import WordCloud

from datasets import load_dataset

from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
    classification_report
)

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

sns.set(style="whitegrid")

# NLTK downloads (run once)
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("wordnet")

# =============================================================================
# 1. DATASET COLLECTION & PREPARATION
# =============================================================================

# Using TweetEval 'sentiment' as real-world dataset
dataset = load_dataset("tweet_eval", "sentiment")

# Label mapping: 0 = negative, 1 = neutral, 2 = positive
label_map = {0: "negative", 1: "neutral", 2: "positive"}

# Convert to DataFrame
train_df = pd.DataFrame({
    "text": dataset["train"]["text"],
    "label": dataset["train"]["label"],
})
val_df = pd.DataFrame({
    "text": dataset["validation"]["text"],
    "label": dataset["validation"]["label"],
})
test_df = pd.DataFrame({
    "text": dataset["test"]["text"],
    "label": dataset["test"]["label"],
})

train_df["label_name"] = train_df["label"].map(label_map)
val_df["label_name"] = val_df["label"].map(label_map)
test_df["label_name"] = test_df["label"].map(label_map)

print("Train size:", len(train_df), "Val size:", len(val_df), "Test size:", len(test_df))
print("\nTrain label distribution:")
print(train_df["label_name"].value_counts())

from sklearn.utils import resample

def balance_df(df, label_col="label"):
    counts = df[label_col].value_counts()
    max_count = counts.max()
    balanced_parts = []
    for lab, cnt in counts.items():
        part = df[df[label_col] == lab]
        if cnt < max_count:
            part = resample(
                part,
                replace=True,
                n_samples=max_count,
                random_state=42
            )
        balanced_parts.append(part)
    return pd.concat(balanced_parts).sample(frac=1, random_state=42).reset_index(drop=True)

train_df = balance_df(train_df, "label")
print("\nBalanced train label distribution:")
print(train_df["label_name"].value_counts())

# =============================================================================
# 2. EXPLORATORY DATA ANALYSIS (EDA) + PREPROCESSING
# =============================================================================

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    """Lowercase, remove mentions/hashtags/urls/punct, tokenize, remove stopwords, lemmatize."""
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"@[A-Za-z0-9_]+", "", text)
    text = re.sub(r"#[A-Za-z0-9_]+", "", text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    tokens = word_tokenize(text)
    tokens = [
        lemmatizer.lemmatize(w)
        for w in tokens
        if w not in stop_words and len(w) > 1
    ]
    return " ".join(tokens)

# Clean text columns
train_df["clean_text"] = train_df["text"].apply(preprocess_text)
val_df["clean_text"]   = val_df["text"].apply(preprocess_text)
test_df["clean_text"]  = test_df["text"].apply(preprocess_text)

# ---- Class distribution plot ----
plt.figure(figsize=(6, 4))
sns.countplot(x="label_name", data=train_df,
              order=["negative", "neutral", "positive"])
plt.title("Class Distribution (Balanced Train Set)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ---- Sentence length distribution  ----
train_df["length"] = train_df["clean_text"].str.split().str.len()

plt.figure(figsize=(6, 4))
sns.histplot(train_df["length"], bins=40, kde=True)
plt.title("Sentence Length Distribution (Words)")
plt.xlabel("Length (words)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# ---- WordClouds per class ----
def plot_wordcloud_for_class(df, label_name, max_words=200):
    text = " ".join(df.loc[df["label_name"] == label_name, "clean_text"])
    if not text.strip():
        print(f"No text for label {label_name}")
        return
    wc = WordCloud(width=800, height=400,
                   background_color="white",
                   max_words=max_words)
    wc.generate(text)
    plt.figure(figsize=(8, 4))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"WordCloud for {label_name} tweets")
    plt.show()

for lab in ["negative", "neutral", "positive"]:
    plot_wordcloud_for_class(train_df, lab)

# ---- Bigram analysis  ----
vectorizer_bi = CountVectorizer(ngram_range=(2, 2), max_features=50)
X_bi = vectorizer_bi.fit_transform(train_df["clean_text"])
bigrams = vectorizer_bi.get_feature_names_out()
counts = np.asarray(X_bi.sum(axis=0)).flatten()
bigram_freq = pd.DataFrame({"bigram": bigrams, "count": counts}) \
    .sort_values("count", ascending=False)

print("\nTop 10 bigrams:")
print(bigram_freq.head(10))

plt.figure(figsize=(8, 4))
sns.barplot(
    data=bigram_freq.head(10),
    x="count",
    y="bigram"
)
plt.title("Top 10 Bigrams (Train)")
plt.xlabel("Count")
plt.ylabel("Bigram")
plt.tight_layout()
plt.show()

# =============================================================================
# 3. SENTIMENT CLASSIFICATION
# =============================================================================

# ---- TF-IDF features ----
tfidf = TfidfVectorizer(
    max_features=20000,
    ngram_range=(1, 2),
    min_df=5
)

X_train = tfidf.fit_transform(train_df["clean_text"])
X_val   = tfidf.transform(val_df["clean_text"])
X_test  = tfidf.transform(test_df["clean_text"])

y_train = train_df["label"].values
y_val   = val_df["label"].values
y_test  = test_df["label"].values

# Helper function for model training & evaluation
def evaluate_model(name, model, X_tr, y_tr, X_te, y_te):
    """Fit model, predict on test set, print metrics & confusion matrix, return metrics dict."""
    # Cross-validation on training set
    cv_scores = cross_val_score(model, X_tr, y_tr, cv=3, scoring="f1_macro")
    print("\n" + "="*70)
    print(f"{name} - Cross-Validation F1 (macro): {cv_scores}")
    print(f"{name} - Mean CV F1 (macro): {cv_scores.mean():.4f}")

    # Fit on full training set and evaluate on test
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)

    acc = accuracy_score(y_te, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_te, y_pred, average="macro", zero_division=0
    )
    cm = confusion_matrix(y_te, y_pred)

    print(f"\n{name} - Test Metrics")
    print("-" * 70)
    print(f"Accuracy:       {acc:.4f}")
    print(f"Precision (macro): {prec:.4f}")
    print(f"Recall (macro):    {rec:.4f}")
    print(f"F1-score (macro):  {f1:.4f}")

    print("\nClassification Report:")
    print(classification_report(
        y_te,
        y_pred,
        target_names=["negative", "neutral", "positive"]
    ))

    plt.figure(figsize=(5, 4))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["negative", "neutral", "positive"],
        yticklabels=["negative", "neutral", "positive"]
    )
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    return {
        "model": name,
        "accuracy": acc,
        "precision_macro": prec,
        "recall_macro": rec,
        "f1_macro": f1,
        "cv_f1_mean": cv_scores.mean()
    }

results = []

# ---- Traditional Model 1: Linear SVM ----
svm_model = LinearSVC(class_weight="balanced", random_state=42)
svm_metrics = evaluate_model("LinearSVC (baseline)", svm_model, X_train, y_train, X_test, y_test)
results.append(svm_metrics)

# ---- Traditional Model 2: Multinomial Naive Bayes ----
nb_model = MultinomialNB()
nb_metrics = evaluate_model("MultinomialNB", nb_model, X_train, y_train, X_test, y_test)
results.append(nb_metrics)

# ---- Traditional Model 3: Logistic Regression ----
logreg_model = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    multi_class="multinomial",
    solver="lbfgs",
    n_jobs=-1
)
logreg_metrics = evaluate_model("LogisticRegression (baseline)", logreg_model, X_train, y_train, X_test, y_test)
results.append(logreg_metrics)

# ---- Hyperparameter Tuning: Linear SVM ----
print("\n\n===== Hyperparameter Tuning: LinearSVC =====")
svm_param_grid = {"C": [0.5, 1.0, 2.0]}
svm_base = LinearSVC(class_weight="balanced", random_state=42)

svm_grid = GridSearchCV(
    svm_base,
    param_grid=svm_param_grid,
    cv=3,
    scoring="f1_macro",
    n_jobs=-1,
    verbose=1
)
svm_grid.fit(X_train, y_train)
print("Best SVM params:", svm_grid.best_params_)
print("Best SVM CV score (F1 macro):", svm_grid.best_score_)

svm_tuned = svm_grid.best_estimator_
svm_tuned_metrics = evaluate_model("LinearSVC (Tuned)", svm_tuned, X_train, y_train, X_test, y_test)
results.append(svm_tuned_metrics)

# ---- Hyperparameter Tuning: Logistic Regression ----
print("\n\n===== Hyperparameter Tuning: LogisticRegression =====")
logreg_param_grid = {"C": [0.5, 1.0, 2.0]}
logreg_base = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    multi_class="multinomial",
    solver="lbfgs",
    n_jobs=-1
)

logreg_grid = GridSearchCV(
    logreg_base,
    param_grid=logreg_param_grid,
    cv=3,
    scoring="f1_macro",
    n_jobs=-1,
    verbose=1
)
logreg_grid.fit(X_train, y_train)
print("Best LogReg params:", logreg_grid.best_params_)
print("Best LogReg CV score (F1 macro):", logreg_grid.best_score_)

logreg_tuned = logreg_grid.best_estimator_
logreg_tuned_metrics = evaluate_model("LogisticRegression (Tuned)", logreg_tuned, X_train, y_train, X_test, y_test)
results.append(logreg_tuned_metrics)

# ---- Pretrained Model: RoBERTa ----
print("\n\n===== Pretrained RoBERTa (cardiffnlp/twitter-roberta-base-sentiment-latest) =====")
roberta_model_name = "cardiffnlp/twitter-roberta-base-sentiment-latest"

tokenizer_rb = AutoTokenizer.from_pretrained(roberta_model_name)
roberta_model = AutoModelForSequenceClassification.from_pretrained(roberta_model_name)
roberta_pipeline = TextClassificationPipeline(
    model=roberta_model,
    tokenizer=tokenizer_rb,
    return_all_scores=False,
    truncation=True,
    max_length=128,
    device=-1
)

sample_size = 1000
test_sample = test_df.sample(min(sample_size, len(test_df)), random_state=42).reset_index(drop=True)

label_str_to_id = {"negative": 0, "neutral": 1, "positive": 2}

y_true_rb = test_sample["label"].values
pred_labels_rb = []

for text in test_sample["text"]:
    pred = roberta_pipeline(text)[0]["label"].lower()
    pred_labels_rb.append(label_str_to_id[pred])

y_pred_rb = np.array(pred_labels_rb)

acc_rb = accuracy_score(y_true_rb, y_pred_rb)
prec_rb, rec_rb, f1_rb, _ = precision_recall_fscore_support(
    y_true_rb, y_pred_rb, average="macro", zero_division=0
)
cm_rb = confusion_matrix(y_true_rb, y_pred_rb)

print("\nRoBERTa - Test Metrics (sample)")
print("-" * 70)
print(f"Accuracy:         {acc_rb:.4f}")
print(f"Precision (macro): {prec_rb:.4f}")
print(f"Recall (macro):    {rec_rb:.4f}")
print(f"F1-score (macro):  {f1_rb:.4f}")

plt.figure(figsize=(5, 4))
sns.heatmap(
    cm_rb,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["negative", "neutral", "positive"],
    yticklabels=["negative", "neutral", "positive"]
)
plt.title("Confusion Matrix - RoBERTa (Sample)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()

results.append({
    "model": "RoBERTa (pretrained, sample)",
    "accuracy": acc_rb,
    "precision_macro": prec_rb,
    "recall_macro": rec_rb,
    "f1_macro": f1_rb,
    "cv_f1_mean": np.nan
})

# =============================================================================
# 4. FINAL RESULTS TABLE
# =============================================================================

results_df = pd.DataFrame(results)
print("\n\n===== SUMMARY OF MODELS =====")
print(results_df.sort_values("f1_macro", ascending=False).reset_index(drop=True))
```

```python
# Write your code here
# ===================== QUESTION 2: TEXT CLASSIFICATION =======================
# Imports & Setup
import warnings
warnings.filterwarnings("ignore")

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

sns.set(style="whitegrid")

# NLTK resources
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("wordnet")
nltk.download("stopwords")

stop_words = set(stopwords.words("english"))
lemm = WordNetLemmatizer()

# ===================== 1. LOAD DATA  =====================

def load_stsa_file(path):
    """
    STSA format: each line => <label><space><text...>
    label: 1 = positive, 0 = negative
    """
    labels = []
    texts = []
    with open(path, "r", encoding="utf8") as f:
        for line in f:
            parts = line.strip().split(" ", 1)
            if len(parts) == 2:
                label, text = parts
                labels.append(int(label))
                texts.append(text)
    return pd.DataFrame({"label": labels, "text": texts})

train_df = load_stsa_file("/content/stsa-train.txt")
test_df  = load_stsa_file("/content/stsa-test.txt")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print(train_df.head())

# ===================== 2. EDA on TRAIN & TEST ===============================

def plot_class_dist(df, name):
    plt.figure(figsize=(5, 4))
    sns.countplot(x=df["label"])
    plt.title(f"{name} Label Distribution (0=neg, 1=pos)")
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

plot_class_dist(train_df, "Train")
plot_class_dist(test_df, "Test")

# Text length distributions
train_df["length"] = train_df["text"].apply(lambda x: len(str(x).split()))
test_df["length"]  = test_df["text"].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(6, 4))
sns.histplot(train_df["length"], bins=50, kde=True)
plt.title("Train Text Length Distribution")
plt.xlabel("Number of Tokens")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(test_df["length"], bins=50, kde=True)
plt.title("Test Text Length Distribution")
plt.xlabel("Number of Tokens")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# ===================== 3. PREPROCESSING =====================================

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    tokens = word_tokenize(text)
    tokens = [lemm.lemmatize(w) for w in tokens if w not in stop_words and len(w) > 2]
    return " ".join(tokens)

train_df["clean"] = train_df["text"].apply(clean_text)
test_df["clean"]  = test_df["text"].apply(clean_text)

print("\nSample cleaned text:")
print(train_df[["text", "clean"]].head())

# ===================== 4. TRAIN SPLIT ===============================

X_full = train_df["clean"]
y_full = train_df["label"]

# 80% for training (with 10-fold CV inside), 20% as held-out validation
X_train_text, X_val_text, y_train, y_val = train_test_split(
    X_full,
    y_full,
    test_size=0.2,
    random_state=42,
    stratify=y_full
)

X_test_text = test_df["clean"]
y_test = test_df["label"]

print("\nTrain size (80% of original train):", X_train_text.shape[0])
print("Val size (20% of original train):  ", X_val_text.shape[0])
print("Test size:                         ", X_test_text.shape[0])

# ===================== 5. TF-IDF FEATURES ===================================

tfidf = TfidfVectorizer(
    max_features=15000,
    ngram_range=(1, 2),
    min_df=3
)

X_train_tfidf = tfidf.fit_transform(X_train_text)
X_val_tfidf   = tfidf.transform(X_val_text)
X_test_tfidf  = tfidf.transform(X_test_text)

print("\nTF-IDF shapes:")
print("X_train:", X_train_tfidf.shape)
print("X_val:  ", X_val_tfidf.shape)
print("X_test: ", X_test_tfidf.shape)

# ===================== 6. 10-FOLD CROSS-VALIDATION SETUP ====================

kf = KFold(n_splits=10, shuffle=True, random_state=42)

def cv_f1(model, X, y):
    """10-fold cross-validated F1 (binary)."""
    scores = cross_val_score(model, X, y, cv=kf, scoring="f1")
    return scores.mean(), scores

# ===================== 7. MODEL TRAINING & EVALUATION =======================

results = []

def evaluate_model(name, model, Xtr, ytr, Xte, yte):
    """
    - Runs 10-fold CV on training split (Xtr, ytr)
    - Trains on full training split
    - Evaluates on TEST (Xte, yte)
    - Prints metrics and returns a dict
    """
    cv_mean, cv_scores = cv_f1(model, Xtr, ytr)
    print("\n" + "="*70)
    print(f"{name} : 10-Fold CV F1 on training (80% split)")
    print("Fold F1 scores:", np.round(cv_scores, 4))
    print("Mean CV F1:", round(cv_mean, 4))

    # Train on full training split
    model.fit(Xtr, ytr)
    preds = model.predict(Xte)

    acc  = accuracy_score(yte, preds)
    prec = precision_score(yte, preds)
    rec  = recall_score(yte, preds)
    f1   = f1_score(yte, preds)

    cm = confusion_matrix(yte, preds)

    print("\nTest Performance:")
    print("Accuracy :", round(acc, 4))
    print("Precision:", round(prec, 4))
    print("Recall   :", round(rec, 4))
    print("F1 Score :", round(f1, 4))
    print("\nClassification report (Test):")
    print(classification_report(yte, preds, digits=4))

    plt.figure(figsize=(4.5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {name} (Test)")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    results.append({
        "model": name,
        "cv_f1_mean": cv_mean,
        "test_accuracy": acc,
        "test_precision": prec,
        "test_recall": rec,
        "test_f1": f1
    })

# ---- 1) SVM (LinearSVC) ----
svm = LinearSVC()
evaluate_model("SVM (LinearSVC)", svm, X_train_tfidf, y_train, X_test_tfidf, y_test)

# ---- 2) KNN ----
knn = KNeighborsClassifier(n_neighbors=5)
evaluate_model("KNN (k=5)", knn, X_train_tfidf, y_train, X_test_tfidf, y_test)

# ---- 3) Decision Tree ----
dt = DecisionTreeClassifier(random_state=42)
evaluate_model("Decision Tree", dt, X_train_tfidf, y_train, X_test_tfidf, y_test)

# ---- 4) Random Forest ----
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
evaluate_model("Random Forest", rf, X_train_tfidf, y_train, X_test_tfidf, y_test)

# ---- 5) XGBoost ----
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=42
)
evaluate_model("XGBoost", xgb, X_train_tfidf, y_train, X_test_tfidf, y_test)

# ===================== 8. SUMMARY TABLE =====================================

results_df = pd.DataFrame(results)
print("\n\n===== SUMMARY OF MODELS (sorted by Test F1) =====")
print(results_df.sort_values("test_f1", ascending=False).reset_index(drop=True))
```

```python
# ===================== QUESTION 3: TEXT CLUSTERING ===========================
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from gensim.models import Word2Vec

sns.set(style="whitegrid")
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

# ============ 1. Load dataset & basic EDA =============
df = pd.read_csv("test.csv", header=None, names=["Rating", "Summary", "Review"])

print("Dataset shape:", df.shape)
print(df.columns)
print(df.head())

# Use the full review text column
df = df.dropna(subset=["Review"]).reset_index(drop=True)
df = df.sample(min(len(df), 5000), random_state=42)

print("After dropna/sample, shape:", df.shape)

df["review_len"] = df["Review"].apply(lambda x: len(str(x)))
print(df["review_len"].describe())

plt.figure(figsize=(6,4))
sns.histplot(df["review_len"], bins=50, kde=True)
plt.title("Review character-length distribution")
plt.xlabel("Length (chars)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ============ 2. Preprocessing =============

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    return " ".join(tokens)

df["clean"] = df["Review"].apply(clean_text)

# For sanity: view some cleaned samples
print(df["clean"].head())

# ============ 3. TF-IDF vectorization + 2D projection =============

tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df["clean"])

# Use PCA for dimensionality reduction before t-SNE (speed / stability)
pca = PCA(n_components=50, random_state=42)
X_pca = pca.fit_transform(X_tfidf.toarray())

tsne = TSNE(n_components=2, perplexity=40, n_iter=500, random_state=42)
X_2d = tsne.fit_transform(X_pca)

# ============ 4. Clustering methods & Visualizations =============

# --- 4.1 K-Means ---
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
k_labels = kmeans.fit_predict(X_tfidf)

plt.figure(figsize=(7,5))
sns.scatterplot(x=X_2d[:,0], y=X_2d[:,1], hue=k_labels, palette="tab10", s=10)
plt.title("K-Means Clustering")
plt.show()

# --- 4.2 DBSCAN ---
dbscan = DBSCAN(eps=0.5, min_samples=10, metric="euclidean")
db_labels = dbscan.fit_predict(X_tfidf)

plt.figure(figsize=(7,5))
sns.scatterplot(x=X_2d[:,0], y=X_2d[:,1], hue=db_labels, palette="tab10", s=10)
plt.title("DBSCAN Clustering (eps=0.5, min_samples=10)")
plt.show()

# --- 4.3 Hierarchical / Agglomerative Clustering ---
agg = AgglomerativeClustering(n_clusters=k)
agg_labels = agg.fit_predict(X_tfidf.toarray())

plt.figure(figsize=(7,5))
sns.scatterplot(x=X_2d[:,0], y=X_2d[:,1], hue=agg_labels, palette="tab10", s=10)
plt.title("Agglomerative Clustering")
plt.show()

# --- 4.4 Word2Vec + K-Means ---
tokenized = [doc.split() for doc in df["clean"]]

w2v_model = Word2Vec(
    sentences=tokenized,
    vector_size=100,
    window=5,
    min_count=5,
    workers=4,
    sg=1
)

def sent_vector(tokens, model, dim=100):
    vecs = [model.wv[w] for w in tokens if w in model.wv]
    if not vecs:
        return np.zeros(dim)
    return np.mean(vecs, axis=0)

w2v_embeddings = np.vstack([sent_vector(tokens, w2v_model, 100) for tokens in tokenized])

kmeans_w2v = KMeans(n_clusters=k, random_state=42)
w2v_labels = kmeans_w2v.fit_predict(w2v_embeddings)

pca_w2v = PCA(n_components=2, random_state=42).fit_transform(w2v_embeddings)

plt.figure(figsize=(7,5))
sns.scatterplot(x=pca_w2v[:,0], y=pca_w2v[:,1], hue=w2v_labels, palette="tab10", s=10)
plt.title("Word2Vec Embeddings + K-Means Clustering")
plt.show()

# Attach K-Means labels and print example reviews from each cluster
df["kmeans_cluster"] = k_labels
for c in range(k):
    print(f"\n--- Sample reviews from K-Means cluster {c} ---")
    print(df.loc[df["kmeans_cluster"] == c, "Review"].head(3).tolist())
```
